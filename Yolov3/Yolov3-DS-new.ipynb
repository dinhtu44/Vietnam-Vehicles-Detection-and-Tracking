```python
for i in [26]:
    video = cv2.VideoCapture('/content/drive/MyDrive/Colab Notebooks/vehicles_detection/train/{}.mp4'.format(i))
    out = cv2.VideoWriter('/content/demo_videos/{}.avi'.format(i),cv2.VideoWriter_fourcc(*'XVID'), int(round(video.get(cv2.CAP_PROP_FPS))), (int(round(video.get(3))),int(round(video.get(4)))))
    j=0
    while(video.isOpened()):
        
        ret, frame = video.read()
        if ret == False or j==40*30:
            break
        if j>=0*30:
            out.write(frame)
            print(i,j)
        j+=1
    video.release()
    cv2.destroyAllWindows()
```
