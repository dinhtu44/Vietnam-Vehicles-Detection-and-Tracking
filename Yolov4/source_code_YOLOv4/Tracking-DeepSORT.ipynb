```python
import cv2
```

**Seperating video into frames**

```python
# quy trình cắt video
# Bộ dữ liệu có gán nhãn từ frames 0-500. Định dạng 60fps (tối đa)
# Nên chọn start từ 20 trở lên để tránh việc tương thích với bộ train

VIDEO_ID =      24    #Tên video muốn tracking
START =         120   #Thời gian bắt đầu tracking (tính bằng (s))
END =           170   #Thời gian kết thúc tracking (tính bằng (s))

for i in [VIDEO_ID]:
    video = cv2.VideoCapture('/content/drive/MyDrive/Train/24.mp4'.format(i)) #đường dẫn truyền video cần phân chia thành các frame
    out = cv2.VideoWriter('/content/drive/MyDrive/demo/{}.avi'.format(i),cv2.VideoWriter_fourcc(*'XVID'), int(round(video.get(cv2.CAP_PROP_FPS))), (int(round(video.get(3))),int(round(video.get(4)))))
    j=0
    while(video.isOpened()):
        
        ret, frame = video.read()
        if ret == False or j==END*30:
            break
        if j>=START*30:
            out.write(frame)
            print(i,j)
        j+=1
    video.release()
    cv2.destroyAllWindows()
```
